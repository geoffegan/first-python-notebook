Merging the propositions and contributors based on their common ID

```python
merge = pd.merge(prop, contribs, on="calaccess_committee_id")
```

```python
merge.info()
```

```python
merge.head()
```

Checking total donations from large donors

```python
merge.amount.sum()
```

Checking number of supporting committees v opposing committees

```python
merge.committee_position.value_counts()
```

Singling out supporting and opposing committees together

```python
support = merge[merge.committee_position == "SUPPORT"]
```

```python
oppose = merge[merge.committee_position == "OPPOSE"]
```

Total raised on each side

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / merge.amount.sum()
```

Descending donors list

```python
merge.sort_values("amount", ascending=False)
```

Descending opposition donors list

```python
oppose.sort_values("amount", ascending=False).head()
```

What committees raised the most?

```python
merge.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

Top donors by first name, last name (Seems to not be picking up corporate donors without first names?)

```python
merge.groupby(["contributor_firstname", "contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False)
```
